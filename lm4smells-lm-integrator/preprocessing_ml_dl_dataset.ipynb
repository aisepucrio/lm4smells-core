# Imports

```python
import pandas as pd
import numpy as np
from typing import List
```

# Auxiliary codes

```python
def read_csv(path: str):
  return pd.read_csv(path)
```

```python
def write_csv(df_final: pd.DataFrame, file_name: str):
  return df_final.to_csv(file_name, index=False)
```

```python
def len_dataframe(l):
  return len(l)
```

```python
def drop_columns(df: pd.DataFrame, cols_to_drop: List[str]) -> pd.DataFrame:
 return df.drop(columns=cols_to_drop)
```

# Select dataset columns

```python
def select_columns(dataset):
  return dataset.loc[:,['project','PySmell','file_name','method_name','analyse_type','code_metric']]
```

# Mapper metrics dataset

```python

def metrics_mapper(colums):
  metrics=[]
  for m in colums['code_metric']:
    k = eval(m)
    raw_loc= k['raw']['-']['loc']
    raw_lloc = k['raw']['-']['lloc']
    raw_sloc = k['raw']['-']['sloc']
    raw_comments = k['raw']['-']['comments']
    raw_multi = k['raw']['-']['multi']
    raw_blank = k['raw']['-']['blank']
    raw_single_comments = k['raw']['-']['single_comments']
    cc_type = k['cc']['-'][0]['type']
    cc_rank = k['cc']['-'][0]['rank']
    cc_complexity = k['cc']['-'][0]['complexity']
    cc_name = k['cc']['-'][0]['name']
    cc_lineno = k['cc']['-'][0]['lineno']
    cc_col_offset = k['cc']['-'][0]['col_offset']
    cc_endline = k['cc']['-'][0]['endline']
    cc_closures= k['cc']['-'][0]['closures']
    mi_mi = k['mi']['-']['mi']
    mi_rank = k['mi']['-']['rank']
    hal_t_h1 = k['hal']['-']['total']['h1']
    hal_t_h2 = k['hal']['-']['total']['h2']
    hal_t_N1 = k['hal']['-']['total']['N1']
    hal_t_N2 = k['hal']['-']['total']['N2']
    hal_t_vocabulary = k['hal']['-']['total']['vocabulary']
    hal_t_length = k['hal']['-']['total']['length']
    hal_t_calculated_length = k['hal']['-']['total']['calculated_length']
    hal_t_volume = k['hal']['-']['total']['volume']
    hal_t_difficulty = k['hal']['-']['total']['difficulty']
    hal_t_effort = k['hal']['-']['total']['effort']
    hal_t_time = k['hal']['-']['total']['time']
    hal_t_bugs = k['hal']['-']['total']['bugs']
    name = list(k['hal']['-']['functions'].keys())[0]
    hal_func_h1 = k['hal']['-']['functions'][name]['h1']
    hal_func_h2 = k['hal']['-']['functions'][name]['h2']
    hal_func_N1 = k['hal']['-']['functions'][name]['N1']
    hal_func_N2 = k['hal']['-']['functions'][name]['N2']
    hal_func_vocabulary = k['hal']['-']['functions'][name]['vocabulary']
    hal_func_length = k['hal']['-']['functions'][name]['length']
    hal_func_calculated_length = k['hal']['-']['functions'][name]['calculated_length']
    hal_func_volume = k['hal']['-']['functions'][name]['volume']
    hal_func_difficulty = k['hal']['-']['functions'][name]['difficulty']
    hal_func_effort = k['hal']['-']['functions'][name]['effort']
    hal_func_time = k['hal']['-']['functions'][name]['time']
    hal_func_bugs = k['hal']['-']['functions'][name]['bugs']

    r ={"raw_loc":raw_loc,"raw_lloc":raw_lloc,"raw_sloc":raw_sloc,"raw_comments":raw_comments,"raw_multi":raw_multi,
        "raw_blank":raw_blank,"raw_single_comments":raw_single_comments,"cc_type":cc_type,"cc_rank":cc_rank,
        "cc_complexity":cc_complexity,"cc_name":cc_name,"cc_lineno":cc_lineno,"cc_closures":cc_closures,"mi_mi":mi_mi,
        " mi_rank":mi_rank,"hal_t_h1":hal_t_h1,"hal_t_h2":hal_t_h2,"hal_t_N1":hal_t_N1,"hal_t_N2":hal_t_N2,
        "hal_t_vocabulary":hal_t_vocabulary,"hal_t_length":hal_t_length,"hal_t_calculated_length":hal_t_calculated_length,
        "hal_t_volume":hal_t_volume,"hal_t_difficulty":hal_t_difficulty,"hal_t_effort":hal_t_effort,
        "hal_t_time":hal_t_time,"hal_t_bugs":hal_t_bugs,"hal_func_h1":hal_func_h1,"hal_func_h2":hal_func_h2,
        "hal_func_N1":hal_func_N1,"hal_func_N2":hal_func_N2,"hal_func_vocabulary":hal_func_vocabulary,
        "hal_func_length":hal_func_length,"hal_func_calculated_length":hal_func_calculated_length,
        "hal_func_volume":hal_func_volume,"hal_func_difficulty":hal_func_difficulty,"hal_func_effort":hal_func_effort,
        "hal_func_time":hal_func_time,"hal_func_bugs":hal_func_bugs}
    metrics.append(r)
  return metrics
```

```python
long_method = read_csv("qwen2_5_coder_1_5b_long_method_zero_shot_with_ast.csv")
df_new_lm = select_columns(long_method)

result_lm_metrics = metrics_mapper(df_new_lm)

df_lm_metric = pd.DataFrame(result_lm_metrics)
df_lm_metric.head()
```

```python
long_parameter_list = read_csv("qwen2_5_coder_1_5b_long_parameter_list_zero_shot_with_ast.csv")
df_new_lpl = select_columns(long_parameter_list)

result_lpl_metrics = metrics_mapper(df_new_lpl)

df_lpl_metric = pd.DataFrame(result_lpl_metrics)
df_lpl_metric.head()
```

```python
df_r = pd.concat([lm,lpl],axis=0)

# Normalize name/value of the PySmell column
if 'PySmell' not in df_r.columns:
    for c in df_r.columns:
        if c.lower() == 'pysmell':
            df_r = df_r.rename(columns={c: 'PySmell'})
            break
df_r['PySmell'] = pd.to_numeric(df_r['PySmell'], errors='coerce').fillna(0).astype(int)

df_r['analyse_type'] = df_r['analyse_type'].astype(str).str.strip().str.lower()

# Rule:
# - PySmell == 1  -> copy analyse_type (e.g., long-method)
# - PySmell == 0  -> "non-" + analyse_type (e.g., non-long-method)
df_r['smell_label'] = np.where(df_r['PySmell'] == 1,
                               df_r['analyse_type'],
                               'non-' + df_r['analyse_type'])
```

# Drop Code Smell

##### Drop the code smells that are not necessary for the model. For instance, this code is set up to the Long Method smell. To train other models, remove the other code smells.

#### Check the correlation and remove the high-correlated features from the dataset

```python
target = dl_ml_dataset['smell_label']
corr = dl_ml_dataset.corr(numeric_only = True)
threshold = 0.8
#columns = np.full((corr.shape[0],), True,dtype='object')
columns = np.isin(dl_ml_dataset.columns,corr.index)
for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= threshold:
            if columns[j]:
                columns[j] = False
selected_columns = dl_ml_dataset.columns[columns]
high_corr = set(dl_ml_dataset.columns) - set(selected_columns)
df = dl_ml_dataset[selected_columns]
df['smell_label'] = target
df.columns
# Separar target antes da filtragem
```

```python
# Replace special characters with underscore and remove spaces
df.columns = df.columns.str.replace('[^A-Za-z0-9_]+', '_', regex=True).str.strip()
```
